#### The iterrows() Method

Create a Python dictionary object whose keys are the headers of the dataframe created in the read_csv() exercise and values are Python list objects that contain data corresponding to the headers. (Here, use the iterrows method to iterate each row of the dataframe and copy it to a dictionary. However, there is a easier way. Learn how the to_dict() method works by yourself later)

````python
from collections import defaultdict

df_org_dict = defaultdict(list)

for (row_index,row) in df_org.iterrows():
    for header in df_org.columns.values:
        df_org_dict[header].append(row[header])
    
print(df_org_dict)
````

copy the code above to the following cell and add code for plotting curve $\left(x, y1\right)$, $\left(x, y2\right)$ and $\left(x, y3\right)$.

````python
x = np.linspace(-5, 5, num=20)
y1 = np.array([j for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([j ** 3 for j in x])

plt.style.use('classic')
plt.title('Graph of y1 vs x')

plt.plot(x, y1, label='y1 = linear numbers', color='black', linestyle='solid', marker='o',
     markerfacecolor='red', markersize=12)
plt.legend()
plt.text(5, 7, r'$y=x$', fontsize=15)

plt.xlabel('x')
plt.ylabel('y1')
plt.show()
````

````python
plt.style.use('classic')
plt.title('Graph of y2 vs x')

plt.plot(x, y2, label='y2 = sqaures',color='black', linestyle='dotted', marker='o',
     markerfacecolor='blue', markersize=12)
plt.legend()
plt.text(4, 6, r'$y=x^2$', fontsize=15)

plt.xlabel('x')
plt.ylabel('y2')
plt.show()
````

````python
plt.style.use('classic')
plt.title('Graph of y3 vs x')

plt.plot(x, y3, label='y3 = cubes', color='black', linestyle='dashed', marker='o',
     markerfacecolor='green', markersize=12)
plt.legend()
plt.text(4, 6, r'$y=x^3$', fontsize=15)

plt.xlabel('x')
plt.ylabel('y3')
plt.show()
````
